#### Modeling with Statsmodel

```python
import statsmodels.api as sm
```

```python
X = df[['TV']]
X[0:5]
```

```python
X = sm.add_constant(X)
```

```python
X[0:5]
```

```python
y = df[['sales']]
y[0:5]
```

```python
lm = sm.OLS(y, X)
```

```python
model = lm.fit()
```

```python
model.summary()
```

```python
import statsmodels.formula.api as smfa
lm = smfa.ols('sales ~ TV', df)
model1 = lm.fit()
```

```python
model1.summary()
```

```python
model.params
```

```python
model.summary().tables[1]
```

```python
print('f_pvalue', '%.50f' % model.f_pvalue)
```

```python
[intercept, coef] =  model.params
```

```python
formula = 'Sales = ' + str(intercept) + ' + TV * '  + str(coef)
formula
```
